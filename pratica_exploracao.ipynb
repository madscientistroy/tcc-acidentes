```python
import pandas as pd
```

```python
d = pd.read_csv('/content/drive/MyDrive/agrupados_pessoa/acidentes2014.csv', encoding='latin1', sep=";")
```

```python
d.groupby('uf', as_index=False)['id'].count()
```

```python
acidentes2014 = pd.DataFrame(d.groupby('uf')['id'].count()).reset_index()
```

```python
acidentes2014.rename(columns={'id':'acidentes', 'uf':'UF'}, inplace=True)
```

```python
acidentes_mortos = pd.DataFrame(d.groupby('uf')['mortos'].sum()).reset_index()
```

```python
acidentes_mortos
```

```python
acidentes_mortos.rename(columns={'uf':'UF', 'mortos':'MORTOS'}, inplace=True)
```
